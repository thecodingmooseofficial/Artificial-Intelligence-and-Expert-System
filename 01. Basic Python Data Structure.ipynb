<b> Exercise 4:</b> Implement stack using list
<p><b>Hints:</b> The list methods make it very easy to use a list as a stack, where the last element added is the first element retrieved (“last-in, first-out”). To add an item to the top of the stack, use append(). To retrieve an item from the top of the stack, use pop() without an explicit index</p>

```python
Stack = []

# append() o push elements in the Stack
Stack.append('Artificial Intelligence and Expert System')
Stack.append('Data Structure')
Stack.append('Object Oriented Programming 2 (C#)')

print('Initial Stack data --')
print(Stack)

# pop() element from Stack (LIFO)
print('\n*Elements popped from Stack*')
print(Stack.pop())
print(Stack.pop())

print('\nStack data after poping:')
print(Stack)
```

<b> Exercise 5:</b> Implement queue using list
<p><b>Hints:</b> It is also possible to use a list as a queue, where the first element added is the first element retrieved (“first-in, first-out”)</p>

```python
# Initializing a Queue using list
Queue = []

# Adding elements to the Queue
Queue.append('Data Structure')
Queue.append('Object Oriented Programming 2 (C#)')
Queue.append('Artificial Intelligence and Expert System')

print("Initial Queue data --")
print(Queue)

# Removing elements from the Queue
print("\n*Elements deQueued from Queue*")
print(Queue.pop(0))
print(Queue.pop(0))

print("\nQueue after removing data -- ")
print(Queue)
```

# Dictionary
<p style="text-align: justify">Dictionaries are sometimes found in other languages as “associative memories” or “associative arrays”. Unlike sequences, which are indexed by a range of numbers, dictionaries are indexed by keys, which can be any immutable type; strings and numbers can always be keys.</p>

<p style="text-align: justify">It is best to think of a dictionary as a set of key: value pairs, with the requirement that the keys are unique (within one dictionary). A pair of braces creates an empty dictionary: {}. Placing a comma-separated list of key:value pairs within the braces adds initial key:value pairs to the dictionary; this is also the way dictionaries are written on output.</p>

# Tuples and Sets
<b>Exercise 7: </b> See the content from [here](https://docs.python.org/3/tutorial/datastructures.html#tuples-and-sequences) to follow some examples on tuple and set.

```python
# Define a tuple of AI algorithms
ai_algorithms = ("Neural Network", "Decision Tree", "SVM", "Neural Network", "KNN")

# Convert the tuple to a set to remove duplicates
unique_algorithms = set(ai_algorithms)

# Add a new algorithm to the set
unique_algorithms.add("Random Forest")

# Print the original tuple and the unique set
print("Original Tuple of AI algorithms:", ai_algorithms)
print("Unique AI algorithms set:", unique_algorithms)

# Define a tuple of ML model accuracies
model_accuracies = (0.92, 0.87, 0.95, 0.92)

# Convert accuracies to a set to identify unique accuracy values
unique_accuracies = set(model_accuracies)

# Print the accuracies in both tuple and set form
print("Original Tuple of Model Accuracies:", model_accuracies)
print("Unique Model Accuracies Set:", unique_accuracies)
```

* The first part deals with AI algorithms and demonstrates how tuples handle duplicates.
* The second part shows how model accuracies (in tuple form) can be converted to sets to identify unique accuracy values.
